```python
import pandas as pd

# Load data
df = pd.read_csv("02_WeatherEvents_Jan2016-Dec2022.csv", sep=",") # karena belum ada kolom
df
```

```python
# Pastikan kolom tanggal dalam format datetime
df['StartTime(UTC)'] = pd.to_datetime(df['StartTime(UTC)'])

# Filter hanya tahun 2022
df_2022 = df[df['StartTime(UTC)'].dt.year == 2022]

# Simpan ke file baru
df_2022.to_csv("weather_cleaned_2022.csv", index=False, sep=",")
df_2022
```
